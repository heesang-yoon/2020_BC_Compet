### 숭인1동이 관할하는 숭인동

```python
a = '''1~71, 72-3~72-7, 72-10~72-58, 72-60~72-66, 72-73~72-181, 72-188~72-256, 76~156, 
178-1, 178-3, 178-7, 179-4, 179-14~179-20, 179-47~179-72, 179-90~179-95, 181-29~181-32, 181-141~181-152, 
181-168, 181-190, 181-204~181-205, 181-260'''
```

```python
a = a.split(',')
soong1 = []
for source in a:
    source = source.strip(' ')
    source = source.replace('-', '.').replace('\n', '')
    if '~' in source:
        source1 = float(source.split('~')[0])
        source2 = float(source.split('~')[1])
        for value in np.arange(source1, source2, 0.01):
            value = np.around(value, 2)
            soong1.append(value)
            soong1.append(source2)
    else:
        soong1.append(float(source))
```

```python
def givemod(start, end):
    
    dt_index = pd.date_range(start=str(start), end=str(end)) # 주어진 기간들 datetime형식
    dt_list = dt_index.strftime("%Y%m%d").tolist() # 리스트화한다
    
    data_list = [] ## 밑에 딕셔너리들 담을 리스트
    
    for date in dt_list:
        servicekey = 'hxTimG9XZ4vz3x88ws4AP6GhkNHJYQBaYsbywF29wFRW8pZzOD1BMaFVUKBDFimDzqSQ2U%2FUns500cWdo%2BO4EA%3D%3D'
        url = 'http://apis.data.go.kr/B553077/api/open/sdsc/storeListByDate?ServiceKey='
        key = '&key='
        condition = '&indsLclsCd=Q&numOfRows=1000'

        open_api = url + servicekey + key + date + condition
        
        res = requests.get(open_api)
        html = res.content
        soup = bs(html, 'html.parser')
        datas = soup.find_all('item')
        
        for data in datas:
            data_dict = dict() # 한 날짜에 바뀐 가게들을 담는 딕셔너리
            
            data_dict['bizesnm'] = data.select_one('bizesnm').get_text()
            data_dict['brchnm'] = data.select_one('brchnm').get_text()
            data_dict['bizesid'] = data.select_one('bizesid').get_text()
            data_dict['indsmclsnm'] = data.select_one('indsmclsnm').get_text()
            data_dict['ctprvnnm'] = data.select_one('ctprvnnm').get_text()
            data_dict['signgunm'] = data.select_one('signgunm').get_text()
            data_dict['adongnm'] = data.select_one('adongnm').get_text()
            data_dict['lnoadr'] = data.select_one('lnoadr').get_text()
            data_dict['chgGb'] = data.select_one('chgGb').get_text()
            data_dict['chgDt'] = data.select_one('chgDt').get_text()
            
            data_list.append(data_dict)
            
    return data_list
```
